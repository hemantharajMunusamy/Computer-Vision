#### Splitting And Merging Color Channel
```
cv2.split()
cv2.merge()
```

````python
img_nz_bgr = cv2.imread("./NB1/New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)

b,g,r = cv2.split(img_nz_bgr)

plt.figure(figsize=[20,5])

plt.subplot(141)
plt.imshow(r, cmap='gray')
plt.title("red channel")

plt.subplot(142)
plt.imshow(g, cmap='gray')
plt.title("green channel")

plt.subplot(143)
plt.imshow(b, cmap='gray')
plt.title("blue channel")

imgMerged = cv2.merge((r,g,b))
plt.subplot(144)
plt.imshow(imgMerged)
plt.title("Merged Output")
````

#### Converting to different Color Spaces
```
dst = cv2.cvtColor(src, code)
```

#### Changing BGR to RGB

````python
img_nz_rgb = cv2.cvtColor(img_nz_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_nz_rgb)
````

#### Changing to HSV color space

````python
img_nz_hsv = cv2.cvtColor(img_nz_bgr, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img_nz_hsv)

plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title('Hue');
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('Saturation');
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('Value');
plt.subplot(144); plt.imshow(img_nz_rgb); plt.title('Original');
````

#### Modify Channel value

````python
h_new = h + 20
img_merged = cv2.merge((h_new, s, v))
img_nz_rgb_modified = cv2.cvtColor(img_merged, cv2.COLOR_HSV2RGB)

plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h_new, cmap='gray'); plt.title('Hue Modified');
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('Saturation');
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('Value');
plt.subplot(144); plt.imshow(img_nz_rgb_modified); plt.title('Modified Image');
````
