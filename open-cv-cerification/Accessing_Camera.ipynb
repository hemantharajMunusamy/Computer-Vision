## Accessing Camera


```python
import cv2
import sys
```

```python
source = cv2.VideoCapture(0)
```

```python
cv2.namedWindow("Camera Preview", cv2.WINDOW_NORMAL)
```

```python
while cv2.waitKey(1) != 27: # Escape
    has_frame, frame = source.read()
    if not has_frame:
        break
    cv2.imshow("Camera Preview",frame)

source.release()
cv2.destroyWindow("Camera Preview")
```
