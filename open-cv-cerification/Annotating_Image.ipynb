## Image Annotation

<ol>
    <li>Draw Lines</li>
    <li>Draw Circles</li>
    <li>Draw Rectangles</li>
    <li>Add text</li>
</ol>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image
````

````python
image = cv2.imread("./NB3/Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
````

#### Draw line
```
img = cv2.line(img, pt1, pt2, color[,thickness[,lineType[,shift]]]])
```

````python
imgLine = image.copy()

cv2.line(imgLine, (100,100), (400,100), (0,255,255), thickness=5, lineType=cv2.LINE_AA)

plt.imshow(imgLine[:,:,::-1])
````

#### Drawing a Circle
```
img = cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]]])
```

````python
img_circle = image.copy()

cv2.circle(img_circle, (900,500), 100, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA)

plt.imshow(img_circle[:,:,::-1])
````

#### Drawing Rectangle
```
img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]])
```

````python
imageRectangle = image.copy()

cv2.rectangle(imageRectangle, (500,50), (700,650), (255, 0, 255), thickness=5, lineType=cv2.LINE_8)

plt.imshow(imageRectangle[:,:,::-1])
````

#### Adding Text
```
img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]])
```

````python
imgText = image.copy()

cv2.putText(imgText, "Apollo 11 Saturn V Launch, July 16, 1969",(200,700),
            fontFace=cv2.FONT_HERSHEY_PLAIN, fontScale=2.3, color=(0,255,0), thickness=2)

plt.imshow(imgText[:,:,::-1])
````
